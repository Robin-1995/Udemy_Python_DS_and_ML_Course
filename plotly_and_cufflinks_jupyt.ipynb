```python
# plotly is an interactive visualisation library
# Cufflinks is used to connect plotly with pandas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import chart_studio.plotly as py
from plotly import __version__
import cufflinks as cf
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot

init_notebook_mode(connected=True)
cf.go_offline()
%matplotlib inline
```

```python
# Data definition
df = pd.DataFrame(np.random.rand(100,4), columns = ["A", "B", "C", "D"]) #100 rows of 4 columns, randomly distributed numbers
df2 = pd.DataFrame({"Category":["A","B","C"], "Values":[32,43,50]})
```

```python
df.plot()
```

```python
df.iplot(kind="scatter", x="A", y="B", mode="markers")
```

```python
# because df2 is already a categorical dataset we can use categorical plots
df2.iplot(kind="bar", x="Category", y="Values", xTitle = "Category", yTitle="Value")
```

```python
# to do this for non aggregated data we can use aggregate / groupby functions to get it in a suitable form

df.mean().iplot(kind="bar",xTitle = "Category", yTitle="Mean")
```

```python
df.iplot(kind="box")
```

```python
df3 = pd.DataFrame({"x":[1,2,3,4,5], "y":[10,20,30,20,10],"z":[500,400,300,200,100]})
```

```python
# for 3-dimensional data sets we can plot a surface plot
df3.iplot(kind="surface", colorscale="rdylbu")
```

```python
# specifing the column we can represent it as a histogram, if we don't pass in a specific column
# it will plot an overlapping histogram
df["A"].iplot(kind="hist", bins=25)
```

```python
df[["A","B"]]
```

```python
# spread plots are useful for stock and financial data
df[["A","B"]].iplot(kind="spread")
```

```python
# bubble plots varies the size of the bubbles dependent on a third variable (in this case C)
# these sorts of plots are useful for GDP in comparison to population etc 
df.iplot(kind="bubble", x="A", y="B", size="C")
```

```python
# scatter matrix is similar to seabon's pairplot
df.scatter_matrix()
```
